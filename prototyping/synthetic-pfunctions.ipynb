```python
# global settings
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import seaborn
#from scipy.interpolate import griddata
import numpy as np
import random
import math
#from sklearn.metrics import mean_absolute_error, mean_squared_error
#import pandas as pd
```

```python
X = list(np.linspace(0.0, 1.0, num=10))
a1 = random.uniform(0.1, 2.0)
```

## LIN

```python
def f1(x):
    return a1*x

plt.plot(X, [f1(x) for x in x])
```

## POLY

```python
def f2(x):
    return a1*x**2

plt.plot(X, [f2(x) for x in x])
```

## EXP

```python
def f3(x):
    return math.exp(f1(x))

plt.plot(X, [f3(x) for x in x])
```

## EXPABS

```python
def f4(x):
    return math.exp(abs(f1(x)))

plt.plot(X, [f4(x) for x in x])
```

## EXPSQ

```python
def f5(x):
    return math.exp(-(f1(x)**2))

plt.plot(X, [f5(x) for x in x])
```

## GAUSS

```python
def f6(x):
    return math.exp(-(f2(x)))

plt.plot(X, [f6(x) for x in x])
```

## WAVE

```python
def f7(x):
    return math.cos(f1(x)) * f3(x)

plt.plot(X, [f7(x) for x in x])
```

## HAT

```python
def f8(x):
    return f2(x) * f6(x)

plt.plot(X, [f8(x) for x in x])
```
